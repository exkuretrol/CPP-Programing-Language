```c++
int i = 123;

cout.setf(ios::left, ios::adjustfield);

cout << "向左對齊：" << endl;
cout << "i=" << setw(5) << setfill('@')  << i << endl << endl;

cout.setf(ios::showpos);

cout << "於正數前加上+號" << endl;
cout << "i=" << setw(10) << setfill('@')  << i << endl;
```
