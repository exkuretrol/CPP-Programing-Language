```c++
// 08170875-陳家瑋-HW4
// include libraries
#include <iostream>
#include <iomanip>
// bring everything in the namespace into scope
using namespace std;
```

```c++
float f = 20.0f;
float g = 10.834f;

cout.setf(ios::showpoint);

// 設定浮點數顯示的精確度
cout.precision(6);
cout << "顯示小數點之後的值：" << endl;
cout << "f=" << f << ",g=" << g << endl << endl;

cout.precision(8);
cout << "使用小數型態列印：" << endl;
cout << f << "-" << g << "=" << f - g;
```
