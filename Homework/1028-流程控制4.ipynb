```c++
cout << "1~50中的偶數有 ";
int i = 1;
int sos = 0;
while (i <= 50)
{
    if (i % 2 == 0) 
    {
        cout << i << " ";
        sos += i * i;
    }
    i++;
}

cout << endl << "平方值之和為" << sos;
```
