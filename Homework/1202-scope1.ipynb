```c++
// 08170875-陳家瑋-HW1
#include <iostream>
using namespace std;
```

```c++
void adder(int m)
{
    static int sum = 0;
    for (int i = 1; i <= m ; i++)
        sum += i;
    cout << "1+2+...+" << m << "=" << sum;
}
```

```c++
int n;
cin >> n;
adder(n);
```
